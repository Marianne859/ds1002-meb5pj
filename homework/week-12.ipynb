## Pie Chart (3 points)

```python
# Create a pie chart from raw values

slices = [56, 66, 24, 124, 47]
labels = ["US","UK","MX","CA","JP"]
```

```python
import matplotlib.pyplot as plt
labels= ["US","UK","MX","CA","JP"]
slices= [56, 66, 24, 124, 47]
plt.pie(slices, labels=labels, startangle=90, explode=(0, 0.1, 0, 0, 0), autopct='%1.1f%%')
plt.axis('equal')
plt.title('Pie Chart')
plt.show()
```

## Histogram (3 points)

```python
# Create a histogram from random data

import matplotlib.pyplot as plt
import numpy as np

n = 430 + np.random.randn(3000)
```

```python
import matplotlib.pyplot as plt
import numpy as np
n = 430 + np.random.randn(3000)
m = [m for m in range(len(n))]
plt.bar(m, n)
plt.title("Raw Data")
plt.show()

plt.hist(n, bins=20)
plt.title("Histogram")
plt.show()

plt.hist(n, cumulative=True, bins=20)
plt.title("Cumulative Histogram")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

num_points = 50
x = 5 + np.arange(num_points) + np.random.randn(num_points)
y = 10 + np.arange(num_points) + 5 * np.random.randn(num_points)
df = pd.DataFrame({'x': x, 'y': y})
plt.scatter(x, y)
z = np.polyfit(df['x'], df['y'], 1)
p = np.poly1d(z)
plt.plot(df['x'], p(df['x']), "r--")
plt.title('Scatter Plot Example')
plt.show()
```
